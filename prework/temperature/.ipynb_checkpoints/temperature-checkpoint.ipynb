If the sensor detects more than 4 hours with temperatures greater than or equal to 70ºC or any temperature above 80ºC or the average exceeds 65ºC throughout the day, we must give the order to change the cooling system to avoid damaging the processor.

````python
# 3. Items in the list that are greater than 70ºC and print the result
M_70=[]    # lista de elementos
n_70=0     # numero de elementos
for i in range(len(temperatures_C)):
    if temperatures_C[i]>70:
        M_70.append(temperatures_C[i])
        n_70+=1
print ('Numero de elementos mayores que 70: %s' % n_70)
print (M_70)
````

````python
# 5.1 Solve the fault in the sensor by estimating a value
for i in range(len(y)):
    if y[i]==0:
        y[i]=(y[i-1]+y[i+1])/2
# plot
plt.plot(x, y)
plt.axhline(y=70, linewidth=1, color='r')
plt.xlabel('hours')
plt.ylabel('Temperature ºC')
plt.title('Temperatures of our server throughout the day (fixed)')
plt.show()
````

````python
# Bonus: convert the list of ºC to ºFarenheit
def Fahrenheit(C):  
    F=(C*1.8)+32
    return F
    
y_F=[]    
for i in range(len(y)):
    f=Fahrenheit(y[i])
    y_F.append(f)
print ('Lista de temperaturas en ºF:')
print (y_F)
````


## Take the decision
Remember that if the sensor detects more than 4 hours with temperatures greater than or equal to 70ºC or any temperature higher than 80ºC or the average was higher than 65ºC throughout the day, we must give the order to change the cooling system to avoid the danger of damaging the equipment:
* more than 4 hours with temperatures greater than or equal to 70ºC
* some temperature higher than 80ºC
* average was higher than 65ºC throughout the day
If any of these three is met, the cooling system must be changed.


````python
# Print True or False depending on whether you would change the cooling system or not

N=4        
suma=[0]
media_mov=[]
for i, j in enumerate(y,1):
    suma.append(suma[i-1]+j)
    if i>N:
        media_movil=(suma[i]-suma[i-N])/N
        media_mov.append(media_movil)
#print (media_mov)

for i in range(len(media_mov)):   
    if media_mov[i]>65:
        C=True
    elif y[i]>70 and y[i+1]>70 and y[i+2]>70 and y[i+3]>70:
        C=True
    elif y[i]>80 or y[i+1]>80 or y[i+2]>80 or y[i+3]>80:
        C=True
    else:
        C=False
    print (C)
plt.show()
````

**Expected output**: 
```
True
```
